```python
import pandas as pd
import json
import date_cmp as dc
from pathlib import Path
import analysis_logic as ayl
import floor as flo
import os
```

```python
df = pd.read_csv('../data/apt_20y_data.csv')
deal = pd.read_csv('../data/(월) 거래규모별 아파트거래현황.csv')
```

```python
ayl_1 = ayl._load_and_clean_data()
```

```python
df = dc.clean_date_column(df ,'거래일')
day_result = dc.day_stat(df , '거래일', '거래금액')
week_result = dc.week_stat(df , '거래일', '거래금액')
month_result = dc.month_stat(df , '거래일', '거래금액')
year_result = dc.year_stat(df , '거래일', '거래금액')
floor_result = flo.floor_home(df['층'])
monthly_volume_result = ayl_1[0]
monthly_area_result = ayl_1[1]
mdeal = flo.mu_home(deal)

```

```python
monthly_area_r_result = monthly_area_result.loc[:,['시도','시군구','1월_면적','2월_면적','3월_면적','4월_면적','5월_면적',
'6월_면적','7월_면적','8월_면적','9월_면적','10월_면적','11월_면적','12월_면적']]
```

```python

output = {
    "일간": day_result.to_dict(orient='records'),
    "주간": week_result.to_dict(orient='records'),
    "월간": month_result.to_dict(orient='records'),
    "년간": year_result.to_dict(orient='records'),
    "층별": floor_result.to_dict(orient='records'),
    "월별 아파트 거래량": monthly_volume_result.to_dict(orient='records'),
    "월별 아파트 거래 거래량 면적": monthly_area_result.to_dict(orient='records'),
    "월별 아파트 거래 면적": monthly_area_r_result.to_dict(orient='records'),
    "면적": mdeal.to_dict(orient='records')
}

# 각 항목별 JSON 파일 저장 (현재 폴더에 저장)
for key, value in output.items():
    file_name = f"{key}.json"
    with open(file_name, "w", encoding="utf-8") as f:
        json.dump({key: value}, f, ensure_ascii=False, indent=2, default=str)
    print(f"{file_name} 저장 완료 ({len(value)}건)")
```
