```python
from flask import Flask, render_template, send_from_directory
import os

app = Flask(
    __name__,
    template_folder='html',     # HTML 파일 위치
    static_folder=None          # 기본 static 폴더 비활성화 (우리가 직접 처리)
)

# 메인 페이지
@app.route('/')
def index():
    return render_template('index.html')

# 서브 페이지들
@app.route('/<page>')
def show_page(page):
    file_path = os.path.join('html', f'{page}.html')
    if os.path.exists(file_path):
        return render_template(f'{page}.html')
    else:
        return "페이지를 찾을 수 없습니다.", 404

# --- JS 파일 처리 ---
@app.route('/js/<path:filename>')
def serve_js(filename):
    return send_from_directory('js', filename)

# --- PY or JSON 파일 처리 ---
@app.route('/py/<path:filename>')
def serve_py(filename):
    return send_from_directory('py', filename)

# --- 데이터 파일 처리 ---
@app.route('/data/<path:filename>')
def serve_data(filename):
    return send_from_directory('data', filename)

if __name__ == '__main__':
      app.run(
        host = '127.0.0.1',
        port = 5000
    )

```
