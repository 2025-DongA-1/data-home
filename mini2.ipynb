```python
import pandas as pd
apt = pd.read_csv('./data/Apart Deal2020.csv')
apt
```

```python
deck1_5 = 0
d1_1 = []
deck6_10 = 0
d6_1 = []
deck11_15 = 0
d11_1 = []
deck16 = 0
d16_1 = []
for i in apt['층']:
    if 5 >= i >= 1:
        deck1_5 += 1
    elif 10 >= i >= 6:
        deck6_10 += 1
    elif 15 >= i >= 11:
        deck11_15 += 1
    else:
        deck16 += 1
dk = [deck1_5, deck6_10, deck11_15, deck16]
labels = ['1층~5층', '6층~10층', '11층~15층', '16층 이상']
deck = [dk, labels]
deck
```
